```python
# 📌 Poetry Generator using FLAN-T5

# Step 1: Install required libraries
!pip install transformers sentencepiece --quiet

# Step 2: Import libraries
from transformers import T5ForConditionalGeneration, T5Tokenizer
import torch

# Step 3: Load the model and tokenizer
model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Step 4: Define a function to generate poetry
def generate_poem(input_text, max_length=100):
    prompt = f"Write a poem about: {input_text}"
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output = model.generate(
        input_ids,
        max_length=max_length,
        num_beams=5,
        early_stopping=True,
        no_repeat_ngram_size=2
    )
    poem = tokenizer.decode(output[0], skip_special_tokens=True)
    return poem

# Step 5: Enter your input and generate poem
input_text = "a lonely star shining in the vast sky"
poem = generate_poem(input_text)

# Step 6: Display the poem
print("✨ Generated Poem:\n")
print(poem)
```
